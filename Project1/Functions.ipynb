```python
import os
import platform
def print_env_info():
    # print python version
    print(f"Python version: {platform.python_version ()}")
    # print spark version
    print(f"Spark version: {spark.version}" )
    # databricks run time version
    runtime_version=os.environ.get("DATABRICKS_RUNTIME_VERSION","Unknown")
    print(f"Databricks runtime version: {runtime_version}")
    # print current user
    print(f"Current user: {spark.sql('select current_user()').first()[0]}")
    # print current database
    print(f"Current database: {spark.sql('select current_database()').first()[0]}")
# print_env_info()
```
